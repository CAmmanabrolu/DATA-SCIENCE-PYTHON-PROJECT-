## Extracting Tesla Revenue Data Using Webscraping

```python
pip install beautifulsoup4 requests
```

```python
import requests
from bs4 import BeautifulSoup

def get_tesla_revenue():
    # Define the URL for Tesla's Yahoo Finance page
    url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'

    # Send an HTTP GET request to the URL and fetch the page's content
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the page content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the revenue data on the page
        revenue_data = soup.find('span', {'data-reactid': '221'})

        if revenue_data:
            # Extract the revenue value (you may need to do additional parsing depending on the page structure)
            revenue_value = revenue_data.text
            return revenue_value
        else:
            print("Revenue data not found on the page.")
            return None
    else:
        print("Failed to fetch data from the website.")
        return None

def main():
    # Get Tesla's revenue data
    revenue = get_tesla_revenue()

    if revenue:
        print(f"Tesla's Revenue: {revenue}")
    else:
        print("Failed to fetch revenue data.")

if __name__ == "__main__":
    main()
```

## Tesla Stock and Revenue Dashboard -

```python
pip install yfinance beautifulsoup4 matplotlib
```

```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

def fetch_tesla_stock_data():
    ticker = 'TSLA'
    tesla_data = yf.download(ticker, start='2023-01-01', end='2023-07-25')
    return tesla_data

def fetch_tesla_revenue():
    url = 'https://finance.yahoo.com/quote/TSLA/financials?p=TSLA'
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        revenue_data = soup.find('span', {'data-reactid': '221'})
        
        if revenue_data:
            revenue_value = revenue_data.text
            return revenue_value
        else:
            return None
    else:
        return None

def create_tesla_dashboard():
    # Fetch Tesla stock data
    tesla_data = fetch_tesla_stock_data()

    # Fetch Tesla revenue data
    revenue = fetch_tesla_revenue()

    # Create the dashboard
    fig, ax = plt.subplots(2, 1, figsize=(12, 10))

    # Plot Tesla stock price
    ax[0].plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Price', color='b')
    ax[0].set_xlabel('Date')
    ax[0].set_ylabel('Stock Price (USD)')
    ax[0].set_title('Tesla Stock Price Over Time')
    ax[0].grid(True)
    ax[0].legend()

    # Display Tesla revenue data
    if revenue:
        ax[1].text(0.5, 0.5, f"Tesla's Revenue: {revenue}", fontsize=15, ha='center', va='center')
        ax[1].axis('off')
    else:
        ax[1].text(0.5, 0.5, "Failed to fetch revenue data.", fontsize=15, ha='center', va='center')
        ax[1].axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    create_tesla_dashboard()
```
